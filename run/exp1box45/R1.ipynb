```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('R1.csv')
df.head(10)
```

```python
new_df = df.drop(columns=['File'],inplace=False)
df['Reciprocal'].mean()
new_df.head(10)
```

```python
step = 5
out = new_df.groupby(new_df.index//step).mean()
out.head(5)
```

```python
out.insert(0, "Number of Atoms", [200,400,600,800,1000,1500,3000,6000,12000], True)
out.insert(11,"Real(lammps)", [-433.08485,-2603.037,-2095.1588, -1718.9171 , -4075.8242 , -8481.135 , -17929.401 , -9468.357 , -119477.09 ], True)
out.insert(13,"Total(lammps)", [-2521.1267,-7801.7843,-5780.9742,-6661.3 , 8841.0302 ,  -18156.679 , -51041.175 , -83717.998 , 17537.19 ], True)
out.insert(15,"Elong(lammps)", [-2088.0418,-5198.7473,-3685.8154,-4942.3829 , 12916.854 , -9675.5441 , -33111.774 , -74249.641 , 137014.28 ], True)
out.to_csv('MEAN.csv')
out.head(9)
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("PHL 2D Ewald")
ax[0].plot('Number of Atoms','Time', data=out[out['Number of Atoms'] < 1001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].set_ylabel("Time (sec)")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','Time', data=out[out['Number of Atoms'] > 1001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].set_ylabel("Time (sec)")
ax[1].grid()
# Show the graph
plt.show()
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("Integral 2D Ewald")
ax[0].plot('Number of Atoms','Time.1', data=out[out['Number of Atoms'] < 1001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].set_ylabel("Time (sec)")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','Time.1', data=out[out['Number of Atoms'] > 1001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].set_ylabel("Time (sec)")
ax[1].grid()
# Show the graph
plt.show()
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("Error b/w Integral and PHL 2D Ewald (kcal/mol)")
ax[0].plot('Number of Atoms','error', data=out[out['Number of Atoms'] < 1001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].set_ylabel("Error")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','error', data=out[out['Number of Atoms'] > 1001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].set_ylabel("Error")
ax[1].grid()
# Show the graph
plt.show()
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("% Reduction in Time")
ax[0].plot('Number of Atoms','%Reduction in Time(For Integral)', data=out[out['Number of Atoms'] < 1001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','%Reduction in Time(For Integral)', data=out[out['Number of Atoms'] > 1001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].grid()
# Show the graph
plt.show()
```
