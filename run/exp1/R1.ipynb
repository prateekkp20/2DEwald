```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('R1.csv')
df.head(10)
```

```python
new_df = df.drop(columns=['File'],inplace=False)
df['Reciprocal'].mean()
new_df.head(10)
```

```python
step = 5
out = new_df.groupby(new_df.index//step).mean()
out.head(5)
```

```python
out.insert(0, "Number of Atoms", [200,400,600,800,1000,1500,3000,6000,12000,24000,30000,48000], True)
out.insert(11,"Real(lammps)", [1371.7613,4533.7298,1559.9991,-3801.5667,-2199.8632,-364.06602,-2184.4484,-142202.86,144314.75,0,30000,48000], True) #could only calculate the real values in lammps upto 12k atoms
out.insert(13,"Total(lammps)", [ 1785.4417 , 13229.106 , -5951.2131 , -12917.467 , -29671.886 , -32179.811 , -44110.817 , -302903.8 , -22842.976 ,0,0,0 ], True) 
out.insert(15,"Elong(lammps)", [  413.68036 , 8695.3767 , -7511.2122 , -9115.9007 , -27472.023 , -31815.745 , -41926.368 , -160700.93 , -167157.72 ,0,0,0], True) 
out.to_csv('MEAN.csv')
out.head(9)
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("PHL 2D Ewald")
ax[0].plot('Number of Atoms','Time', data=out[out['Number of Atoms'] < 3001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].set_ylabel("Time (sec)")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','Time', data=out[out['Number of Atoms'] > 3001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].set_ylabel("Time (sec)")
ax[1].grid()
# Show the graph
plt.show()
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("Integral 2D Ewald")
ax[0].plot('Number of Atoms','Time.1', data=out[out['Number of Atoms'] < 3001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].set_ylabel("Time (sec)")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','Time.1', data=out[out['Number of Atoms'] > 3001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].set_ylabel("Time (sec)")
ax[1].grid()
# Show the graph
plt.show()
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("Error b/w Integral and PHL 2D Ewald (kcal/mol)")
ax[0].plot('Number of Atoms','error', data=out[out['Number of Atoms'] < 3001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].set_ylabel("Error")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','error', data=out[out['Number of Atoms'] > 3001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].set_ylabel("Error")
ax[1].grid()
# Show the graph
plt.show()
```

```python
fig, ax = plt.subplots(1, 2,figsize=(10,6))
fig.suptitle("% Reduction in Time")
ax[0].plot('Number of Atoms','%Reduction in Time(For Integral)', data=out[out['Number of Atoms'] < 3001], marker='*', alpha=0.4)
ax[0].set_xlabel("Number of Atoms")
ax[0].grid()
 
# And now add something in the second part:
ax[1].plot(  'Number of Atoms','%Reduction in Time(For Integral)', data=out[out['Number of Atoms'] > 3001], marker='*', alpha=0.3)
ax[1].set_xlabel("Number of Atoms")
ax[1].grid()
# Show the graph
plt.show()
```
