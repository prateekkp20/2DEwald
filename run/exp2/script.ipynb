# Threads vs Time (for different box configuration(# of atoms))

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('r1.csv')
new_df = df.drop(columns=['# of Threads'],inplace=False)
df1 = pd.read_csv('r2.csv')
new_df1 = df1.drop(columns=['# of Threads'],inplace=False)
df2 = pd.read_csv('r3.csv')
new_df2 = df2.drop(columns=['# of Threads'],inplace=False)
df3 = pd.read_csv('r4.csv')
new_df3 = df3.drop(columns=['# of Threads'],inplace=False)
```

```python
step = 5
out = new_df.groupby(new_df.index//step).mean()
out1 = new_df1.groupby(new_df1.index//step).mean()
out2 = new_df2.groupby(new_df2.index//step).mean()
out3 = new_df3.groupby(new_df3.index//step).mean()
```

```python
out.insert(0, "Number of Threads", np.linspace(1,24,num=24), True)
out.to_csv('mean.csv')
out1.insert(0, "Number of Threads", np.linspace(1,24,num=24), True)
out1.to_csv('mean1.csv')
out2.insert(0, "Number of Threads", np.linspace(1,24,num=24), True)
out2.to_csv('mean2.csv')
out3.insert(0, "Number of Threads", np.linspace(1,24,num=24), True)
out3.to_csv('mean3.csv')
```

```python
fig, ax = plt.subplots(1, 2,figsize=(13,6))
fig.suptitle("Time Taken to Calculate Reciprocal Sum for 200 Atoms")
ax[0].plot('Number of Threads','Reciprocal PHL', data=out, marker='*', alpha=0.4)
ax[0].set_title("PHL 2D Ewald")
ax[0].set_xlabel("Number of Threads")
ax[0].set_ylabel("Time (sec)")
ax[0].grid()

# And now add something in the second part:
ax[1].plot( 'Number of Threads','Reciprocal Integral', data=out, marker='*', alpha=0.3)
ax[1].set_title("Integral 2D Ewald")
ax[1].set_xlabel("Number of Threads")
ax[1].set_ylabel("Time (sec)")
ax[1].grid()
# Show the graph
plt.show()
```
